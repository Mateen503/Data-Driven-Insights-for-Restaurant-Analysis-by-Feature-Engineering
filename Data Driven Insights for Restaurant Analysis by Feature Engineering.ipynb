identifying most commom price range

```python
price_range_counts=data['Price range'].value_counts()
most_common_price_range=price_range_counts.idxmax()

print("Price Range :",price_range_counts)
print("Most common price:",most_common_price_range)
```

Calculating average Rating for each price range

```python
avg_rating=data.groupby('Price range')['Aggregate rating'].mean()
print("Print Avg rating:",avg_rating)
```

Highest Avg Rating in colors

```python
highest_avg_rating_range = avg_rating.idxmax()
highest_avg_rating = avg_rating.max()

print(f"The price range with the highest average rating is: {highest_avg_rating_range} with an average rating of {highest_avg_rating}")


if 'Color' in data.columns:
    color_for_highest_rating = data[data['Price Range'] == highest_avg_rating_range]['Color'].unique()
    print(f"The color representing the highest average rating is: {color_for_highest_rating}")
```
